```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"


#Drop the following features from consideration - 1st cut
first_cut_df = df_11.drop(columns=[ 'dayod', 'stater' ,
       'hosp', 'hospd', 'rcow', 'citizen', 'health', 'indalg', 'smok100',
       'agesmk', 'smokstat', 'smokhome', 'curruse', 'everuse'])
print(df_11.columns)
df_11.describe(include='all')
```

```python

#Create new categorical variable Age_Bins
first_cut_df['Age_Category'] = pd.cut(first_cut_df['age'], [26, 31, 36, 41, 46, 51, 56, 61, 66, 71, 76, 81,  100], labels=[ '26-30', '31-35','36-40','41-45','46-50','51-55','56-60', '61-65', '66-70', '71-75', '76-80', 'Above 85'])
first_cut_df.groupby(['Age_Category'])['Age_Category'].count()

# create an indicator for cancer from cause113
first_cut_df['canc_temp'] = pd.cut(first_cut_df['cause113'], [0, 18, 43, 200], labels=[0, 1, 2])
first_cut_df.groupby(['canc_temp'])['canc_temp'].count()
first_cut_df['temp0'] = pd.cut(first_cut_df['canc_temp'], [0,1, 2], labels=[1, 0])
first_cut_df.groupby(['temp0'])['temp0'].count()
first_cut_df['cancer'] = first_cut_df.canc_temp*first_cut_df.temp0
first_cut_df['cancer']=first_cut_df['cancer'].fillna(0)
#generating number for cardiovascular illnesses, 2 steps necessary
first_cut_df['cardio_temp'] = pd.cut(first_cut_df['cause113'], [0, 49, 65, 200], labels=[0, 1, 2])
first_cut_df.groupby(['cardio_temp'])['cardio_temp'].count()
first_cut_df['cardio'] = pd.cut(first_cut_df['cardio_temp'], [0, 1, 2], labels=[1,0])
first_cut_df.groupby(['cardio'])['cardio'].count()
first_cut_df['cardio']=first_cut_df['cardio'].fillna(0)
first_cut_df.groupby(['cardio'])['cardio'].count()

```

```python
# First year mortallity by cause


year_df = first_cut_df
year_df = year_df[year_df['age'] >25 ]
year_df.groupby(['Age_Category'])['Age_Category'].count()
year_df.groupby(['inddea'])['inddea'].count()
 # isolating first yaer deaths
year_df['Indicator'] = pd.cut(year_df['follow'], [0, 366, 4183], labels=[1, 0])
year_df.groupby(['Indicator'])['Indicator'].count()
pd.crosstab(year_df.Age_Category, year_df.Indicator, margins=True)
```

```python
# cancer for year 1
year_df['cancer_year1'] = year_df.cancer*year_df.Indicator
year_df.groupby(['cancer_year1'])['cancer_year1'].count()
pd.crosstab(year_df.Age_Category, year_df.cancer_year1, margins=True)
```

```python
# cardio for for year 1
year_df['cardio_year1'] = year_df.cardio*year_df.Indicator
year_df.groupby(['cardio_year1'])['cardio_year1'].count()
pd.crosstab(year_df.Age_Category, year_df.cardio_year1, margins=True)
```

```python
#cause of death for sixth year
year6_df = year_df
year6_df= year6_df[year6_df['follow'] > 1827]
year6_df['age_plus5'] = year6_df['age'] + 5
year6_df['Age_Category6'] = pd.cut(year6_df['age_plus5'], [ 26, 31, 36, 41, 46, 51, 56, 61, 66, 71, 76, 81,  100], labels=[ '26-30', '31-35','36-40','41-45','46-50','51-55','56-60', '61-65', '66-70', '71-75', '76-80', 'Above 85'])
year6_df.groupby(['Age_Category6'])['Age_Category6'].count()
year6_df.groupby(['inddea'])['inddea'].count()
# creating indicator variable to pick out only the sith year death
year6_df['Indicator6'] = pd.cut(year6_df['follow'], [1827, 2192 , 4183], labels=[1, 0])
pd.crosstab(year6_df.Age_Category6, year6_df.Indicator6, margins=True)
```

```python
# cancer for year 6
year6_df['cardio_year6'] = year6_df.cardio*year6_df.Indicator6
year6_df.groupby(['cardio_year6'])['cardio_year6'].count()
```

```python
pd.crosstab(year6_df.Age_Category6, year6_df.cardio_year6, margins=True)
```

```python
year11_df = year_df
year11_df= year_df[year11_df['follow'] > 3652]
year11_df['age_plus10'] = year11_df['age'] + 10
year11_df['Age_Category10'] = pd.cut(year11_df['age_plus10'], [31, 36, 41, 46, 51, 56, 61, 66, 71, 76, 81,  100], labels=['31-36', '36-40','41-45','46-50','51-55','56-60', '61-65', '66-70', '71-75', '76-80', 'Above 85'])
year11_df.groupby(['Age_Category10'])['Age_Category10'].count()
year11_df.groupby(['inddea'])['inddea'].count()
pd.crosstab(year11_df.Age_Category10, year11_df.inddea, margins=True)

```

```python
# creating categories for density
df_11['low_education'] = df_11['Education_Bins_1'] + df_11['Education_Bins_2'] +df_11['Education_Bins_3']
df_11['near_poverty'] = df_11['PovertyPct_Bins_1'] + df_11['PovertyPct_Bins_0']
df_11.groupby(['near_poverty'])['near_poverty'].count()
data_density = df_11.loc[:,[ 'follow', 'IsDead','near_poverty', 'isMale','low_education', 'age', 'ssnyn', 'vt', 'MarriedStatus_Married', 'EmploymentStatus_Employed' ]]

data_density_temp = data_density[data_density['IsDead'] > 0]
import matplotlib.pyplot as plt
import seaborn as sns

# matplotlib histogram
plt.hist(data_density_temp['age'], color = 'red', edgecolor = 'black',
         bins = int(180/5))

# Add labels
plt.title('Histogram of Age of Death')
plt.xlabel('Age')
plt.ylabel('Death')
plt.savefig('death_pop.png', transparent = True)
```

```python

data_density1 = data_density[data_density['IsDead'] > 0]
data_density1 = data_density1[data_density1['age'] > 50]
import matplotlib.pyplot as plt
import seaborn as sns

# matplotlib histogram
plt.hist(data_density1['age'], color = 'red', edgecolor = 'black',
         bins = int(180/5))

# Add labels
plt.title('Histogram of Age of Death')
plt.xlabel('Age')
plt.ylabel('Death')
```

```python
#Create new variable Sex_Id
data_density1['Sex_ID'] = pd.cut(data_density1['isMale'], [-1, 0, 1], labels=['Women', 'Men'])
data_density1.groupby(['Sex_ID'])['Sex_ID'].count()
# List gender
gender1 = ['Women', 'Men']

# Iterate gender
for gender1 in gender1:
    subset = data_density1[data_density1['Sex_ID'] == gender1]
    
    # Draw the density plot
    sns.distplot(subset['age'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = gender1)

#format plot
plt.legend(prop={'size': 16}, title = 'Gender')
plt.title('Density Plot with Gender')
plt.xlabel('AGE')
plt.ylabel('Density')
plt.savefig('women_density.png', transparent = True)
```

```python
data_density1['Poverty'] = pd.cut(data_density1['near_poverty'], [-1, 0, 1], labels=['Low and Middle', 'Upper' ])
data_density1.groupby(['Poverty'])['Poverty'].count()

# List income
Income = ['Low and Middle', 'Upper']

# Iterate income
for Income in Income:
    subset = data_density1[data_density1['Poverty'] == Income]
    
    # Draw the density plot
    sns.distplot(subset['age'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = Income)

#format plot
plt.legend(prop={'size': 16}, title = 'Groups')
plt.title('Density Plot with Income Class')
plt.xlabel('AGE')
plt.ylabel('Density')
plt.savefig('income.png', transparent = True)
```

```python

data_density1['Low_Ed'] = pd.cut(data_density1['low_education'], [-1, 0, 1], labels=['College', 'High School' ])
data_density1.groupby(['Low_Ed'])['Low_Ed'].count()
# List, shoud be Education, but income is used here
Income = ['College', 'High School']

# Iterate gender
for Income in Income:
    subset = data_density1[data_density1['Low_Ed'] == Income]
    
    # Draw the density plot
    sns.distplot(subset['age'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = Income)

#format plot
plt.legend(prop={'size': 16}, title = 'Groups')
plt.title('Density Plot by Education')
plt.xlabel('AGE')
plt.ylabel('Density')
plt.savefig('Education.png', transparent = True)
```

```python
data_density1['marriage'] = pd.cut(data_density1['MarriedStatus_Married'], [-1, 0, 1], labels=['Not Married', 'Married' ])
data_density1.groupby(['marriage'])['marriage'].count()

# List, shoud be Education, but income is used here
marry = ['Not Married', 'Married']

# Iterate gender
for marry in marry:
    subset = data_density1[data_density1['marriage'] == marry]
    
    # Draw the density plot
    sns.distplot(subset['age'], hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = marry)

#format plot
plt.legend(prop={'size': 16}, title = 'Groups')
plt.title('Density Plot, Marriatal Status')
plt.xlabel('AGE')
plt.ylabel('Density')
plt.savefig('Marriage.png', transparent = True)
```
